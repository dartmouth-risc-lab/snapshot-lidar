# Comparing the EPC660 camera's dual phase mode with our snapshot technique

```python
import numpy as np
import cv2
from scipy.constants import c, pi
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm, SymLogNorm
from scipy.ndimage import gaussian_filter1d

# uncomment for inline matlab style interactive figures
# %matplotlib widget

%cd ".."
import simulation as sim
import fft_shift as fft_shift
from utils import calculate_snr
from fig_setup import fig_setup

fig_setup()  # comment this if using matplotlib widget because they conflict
```

#### Useful paramters

```python
# data paths and variables
save_figs = False

mod_freq = 60 # MHz
wavelength = 940 # nm
calib_offset = 624.5 # mm
cropX = 0
cropY = 0
gauss_dim='1d' 
phase_variation_axis="x"
using_sigmas = [0, 0.25, 0.5, 0.75, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5, 7, 9]

# from mlx datasheet
unambig_dist = (c*0.5)/(mod_freq *1e6)
coeff_to_rad = unambig_dist / (2 * pi)
```

#### Load data
EVK75027 raw depth data is in mm

```python
with np.load('data/emulation/dist_amp.npz') as gt_data:
    depth_rw = ((gt_data['dist'] - calib_offset) * 1e-3)[cropY:, cropX:] # divide by 1e3 to get meters
    intensity_rw = gt_data['amp'][cropY:, cropX:]
    
with np.load('data/emulation/phase_data.npz') as phase_data:
    I0_rw = phase_data['phase0'][cropY:, cropX:]
    I90_rw = phase_data['phase90'][cropY:, cropX:]
    I180_rw = phase_data['phase180'][cropY:, cropX:]
    I270_rw = phase_data['phase270'][cropY:, cropX:]
    
phase_rw = (depth_rw / coeff_to_rad) - pi # calculate the phase from the depth

# show the real world data
fig_rw, axs = plt.subplots(3,2, figsize=(8,7), layout="constrained")
im = axs[0,0].imshow(intensity_rw, cmap="gray")
axs[0,0].set_title("real world intensity")
fig_rw.colorbar(im, ax=axs[0,0])

im = axs[0,1].imshow(depth_rw, cmap="hsv")
axs[0,1].set_title("real world captured depth (meters)")
fig_rw.colorbar(im, ax=axs[0,1])

im = axs[1,0].imshow(I0_rw, cmap="hsv")
axs[1,0].set_title("phase 0")
fig_rw.colorbar(im, ax=axs[1,0])

im = axs[1,1].imshow(I90_rw, cmap="hsv")
axs[1,1].set_title("phase 90")
fig_rw.colorbar(im, ax=axs[1,1])

im = axs[2,0].imshow(I180_rw, cmap="hsv")
axs[2,0].set_title("phase 180")
fig_rw.colorbar(im, ax=axs[2,0])

im = axs[2,1].imshow(I270_rw, cmap="hsv")
axs[2,1].set_title("phase 270")
fig_rw.colorbar(im, ax=axs[2,1])
```

### Simulate the phase reconstructed by the epc660 and epc635 cameras

```python
composite = np.zeros(phase_rw.shape)
composite[::2, :] = I0_rw[::2, :]
composite[1::2, :] = I90_rw[1::2, :]

recon_phase = np.arctan2(-composite[1::2, :], composite[::2, :]) - pi/2
recon_phase_090 = np.zeros(phase_rw.shape)
recon_phase_090[::2, :] = recon_phase
recon_phase_090[1::2, :] = recon_phase

# simulate the phase reconstructed by the dual mgx mode
comp_mgx_1 = np.zeros(phase_rw.shape)
comp_mgx_2 = np.zeros(phase_rw.shape)

comp_mgx_1[::2, :] = I90_rw[::2, :]
comp_mgx_1[1::2, :] = I180_rw[1::2, :]
comp_mgx_2[::2, :] = I270_rw[::2, :]
comp_mgx_2[1::2, :] = I0_rw[1::2, :]

recon_phase_mgx = np.arctan2(comp_mgx_2[::2,:]-comp_mgx_1[::2,:], comp_mgx_2[1::2,:]-comp_mgx_1[1::2,:]) - pi/2

fig_rw, axs = plt.subplots(1,3, figsize=(13,3), layout="constrained")
im = axs[0].imshow(phase_rw, cmap="hsv", vmin=-pi, vmax=pi)
axs[0].set_title("ground truth")
fig_rw.colorbar(im, ax=axs[0])

im = axs[1].imshow(recon_phase_090, cmap="hsv", vmin=-pi, vmax=pi)
axs[1].set_title("epc660 reconstructed phase")
fig_rw.colorbar(im, ax=axs[1])

im = axs[2].imshow(recon_phase_mgx, cmap="hsv", vmin=-pi, vmax=pi)
axs[2].set_title("mgx reconstructed phase")
fig_rw.colorbar(im, ax=axs[2])
```

### Run a simulation for different sigma values

```python
k = np.deg2rad(90)
cropY = 0

intensity_rw = intensity_rw[cropX:, cropY:]
depth_rw = depth_rw[cropX:, cropY:]
phase_rw = phase_rw[cropX:, cropY:]

snrs_intensity_660 = []
snrs_phase_660 = []
snrs_intensity_635 = []
snrs_phase_635 = []
snrs_intensity = []
snrs_phase = []

recon_intensities = []
recon_phases = []
recon_depths = []

diff_intensities = []
diff_phases = []
diff_depths = []

recon_intensities_635 = []
recon_phases_635 = []
recon_depths_635 = []

recon_intensities_660 = []
recon_phases_660 = []
recon_depths_660 = []

diff_intensities_635 = []
diff_phases_635 = []
diff_depths_635 = []

diff_intensities_660 = []
diff_phases_660 = []
diff_depths_660 = []

for i in range(len(using_sigmas)):
    sigma = using_sigmas[i]
       
    # create a simulation
    fft_shift_sim = fft_shift.FFT_Shift(k, intensity_rw, depth_rw,
                                        omega= 2 * pi * (mod_freq * 1e6))

    # computationally create quad intensity measurements
    fft_shift_sim.create_quad_intensity(method='mlx')
    computed_phases = [fft_shift_sim.I0, fft_shift_sim.I90, fft_shift_sim.I180, fft_shift_sim.I270]

    # compute emulated image intensity
    fft_shift_sim.create_composite_intensity(phase_variation_axis=phase_variation_axis, use_Gaussian=gauss_dim, sigma=sigma, method='mlx') # data slices
    composite_intensity = fft_shift_sim.composite_image

    # calculate the fft shift
    fft_shift_sim.calculate_fft_shifted(phase_variation_axis=phase_variation_axis)

    # reconstruct and scale intensity, phase, depth from fft shift
    recon_intensity = 2 * fft_shift_sim.reconstruct_intensity()
    recon_phase = fft_shift_sim.reconstruct_phase() - (pi/2.0)
    recon_depth = ((c * (recon_phase + pi)) / (4 * pi * mod_freq * 1e6))
    
    recon_intensities.append(recon_intensity)
    recon_phases.append(recon_phase)
    
    diff_intensities.append(intensity_rw - recon_intensity)
    diff_phases.append(phase_rw - recon_phase)
    
    snrs_intensity.append(calculate_snr(intensity_rw, recon_intensity))
    snrs_phase.append(calculate_snr(phase_rw, recon_phase))

    # Compare with epc660 camera reconstruction
    filter_axis = 1 if phase_variation_axis == 'x' else 0
    if sigma != 0:
        I0_cur = gaussian_filter1d(I0_rw, sigma, axis=filter_axis, mode='constant')
        I90_cur = gaussian_filter1d(I90_rw, sigma, axis=filter_axis, mode='constant')
        I180_cur = gaussian_filter1d(I180_rw, sigma, axis=filter_axis, mode='constant')
        I270_cur = gaussian_filter1d(I270_rw, sigma, axis=filter_axis, mode='constant')
    else:
        I0_cur = I0_rw
        I90_cur = I90_rw
        I180_cur = I180_rw
        I270_cur = I270_rw
        
    recon_phase = np.arctan2(-I90_cur[1::2, :], I0_cur[::2, :]) - pi/2
    recon_phase_660 = np.zeros(phase_rw.shape)
    recon_phase_660[::2, :] = recon_phase
    recon_phase_660[1::2, :] = recon_phase
    
    # Compare with epc635 camera reconstruction dual MGX mode
    recon_phase = np.arctan2(I270_cur[::2, :]-I90_cur[::2,:],I0_cur[1::2, :]-I180_cur[1::2,:]) - pi/2
    recon_phase_635 = np.zeros(phase_rw.shape)
    recon_phase_635[::2, :] = recon_phase
    recon_phase_635[1::2, :] = recon_phase
        
    recon_phases_660.append(recon_phase_660)
    recon_phases_635.append(recon_phase_635)

    diff_phases_660.append(phase_rw - recon_phase_660)
    diff_phases_635.append(phase_rw - recon_phase_635)
    
    snrs_phase_660.append(calculate_snr(phase_rw, recon_phase_660))
    snrs_phase_635.append(calculate_snr(phase_rw, recon_phase_635))
```

```python
print("SNR for 660 camera: ", snrs_phase_660[4])
print("SNR for 635 dual phase mose: ", snrs_phase_635[4])
print("Our method: ", snrs_phase[4])
```

#### Plot

```python
fig_recon, axs = plt.subplots(3,2, figsize=(5,6), layout="constrained")
opt_sigma_660 = np.argmax(snrs_phase_660)
opt_sigma_635 = np.argmax(snrs_phase_635)
opt_sigma_sp = np.argmax(snrs_phase)

img = axs[0,0].imshow(recon_phases_660[opt_sigma_660], cmap="hsv", vmin = -pi, vmax=pi)
axs[0,0].set_title("Reconstruction")
axs[0,0].axis('off')

img = axs[0,1].imshow(diff_phases_660[opt_sigma_660], cmap="hsv", vmin = -pi/2, vmax=pi/2)
axs[0,1].set_title("Error")
axs[0,1].axis('off')

img = axs[1,0].imshow(recon_phases_635[opt_sigma_635], cmap="hsv", vmin = -pi, vmax=pi)
axs[1,0].axis('off')

img = axs[1,1].imshow(diff_phases_635[opt_sigma_635], cmap="hsv", vmin = -pi/2, vmax=pi/2)
axs[1,1].axis('off')

img = axs[2,0].imshow(recon_phases[opt_sigma_sp], cmap="hsv", vmin = -pi, vmax=pi)
axs[2,0].axis('off')
cbar=fig_recon.colorbar(img, ax=axs[2,0], location='bottom', shrink = 0.8)
cbar.ax.tick_params(labelsize=10)

img = axs[2,1].imshow(diff_phases[opt_sigma_sp], cmap="hsv", vmin = -pi/2, vmax=pi/2)
axs[2,1].axis('off')
cbar=fig_recon.colorbar(img, ax=axs[2,1], location='bottom', shrink = 0.8)
cbar.ax.tick_params(labelsize=10)

if save_figs:
    fig_recon.savefig("figures/CVPRfigures/660camera_vs_snapshot_recons.pdf", bbox_inches = 'tight')
```
