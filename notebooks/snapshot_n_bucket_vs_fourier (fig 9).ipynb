```python
import numpy as np
import cv2
from scipy.constants import c, pi
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm, SymLogNorm
from scipy.ndimage import gaussian_filter1d

# uncomment for inline matlab style interactive figures
# %matplotlib widget

%cd "../"
import simulation as sim
import fft_shift as fft_shift
from utils import calculate_snr
import fig_setup

fig_setup.fig_setup() # comment this if using matplotlib widget because they conflict
```

#### Useful parameters

```python
save_figs = False

mod_freq = 60 # MHz
wavelength = 940 # nm
calib_offset = 624.5 # mm
cropX = 0
cropY = 0
gauss_dim='1d' 
phase_variation_axis="x"
using_sigmas = [0, 0.25, 0.5, 0.75, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5]

unambig_dist = (c*0.5)/(mod_freq *1e6)
coeff_to_rad = unambig_dist / (2 * pi)
```

#### Load data

```python
with np.load('data/emulation/dist_amp.npz') as gt_data:
    depth_rw = ((gt_data['dist'] - calib_offset) * 1e-3)[cropY:, cropX:] # divide by 1e3 to get meters
    intensity_rw = gt_data['amp'][cropY:, cropX:]
    
with np.load('data/emulation/phase_data.npz') as phase_data:
    I0_rw = phase_data['phase0'][cropY:, cropX:]
    I90_rw = phase_data['phase90'][cropY:, cropX:]
    I180_rw = phase_data['phase180'][cropY:, cropX:]
    I270_rw = phase_data['phase270'][cropY:, cropX:]

phase_rw = (depth_rw / coeff_to_rad) - pi # calculate the phase from the depth
```

```python
""" Convert phase variation increment (in degrees) to number of rows/columns cropped
"""
def degrees2crop(degree, size=640):
    mod = size % (360 / degree)
    return int((360 / degree) + mod) if mod != 0 else 0 
```

### Extended visualization

```python
k = np.arange(3,11, 1)
degrees = np.array(list(map(lambda x : 360/x, k)))
degrees = np.insert(degrees, 1, 360/3.5)

snrs_intensity_bucket = np.empty((len(degrees), len(using_sigmas)))
snrs_phase_bucket = np.empty((len(degrees), len(using_sigmas)))
snrs_intensity = np.empty((len(degrees), len(using_sigmas)))
snrs_phase = np.empty((len(degrees), len(using_sigmas)))

fig_recon, axs = plt.subplots(len(degrees),2, figsize=(6,15), layout="constrained")

for index, degree in enumerate(degrees):
    k = np.deg2rad(degree)
    if index==1:
        cropY=10
    else:
        cropY =  degrees2crop(degree) #int(640 % (360 / degree))
        
    # print("cropY for k={} : ".format(2*pi/k), cropY)
    
    intensity_rw_crop = intensity_rw[cropX:, cropY:]
    depth_rw_crop = depth_rw[cropX:, cropY:]
    phase_rw_crop = phase_rw[cropX:, cropY:]

    recon_intensities = []
    recon_phases = []
    recon_depths = []

    diff_intensities = []
    diff_phases = []
    diff_depths = []
    
    n_bucket_images = []
    recon_intensities_bucket = []
    recon_phases_bucket = []
    recon_depths_bucket = []

    diff_intensities_bucket = []
    diff_phases_bucket = []
    diff_depths_bucket = []
    
    for i in range(len(using_sigmas)):
        sigma = using_sigmas[i]

        # create a simulation
        fft_shift_sim = fft_shift.FFT_Shift(k,
                                            intensity_rw_crop, depth_rw_crop,
                                            omega= 2 * pi * (mod_freq * 1e6))

        # computationally create quad intensity measurements
        fft_shift_sim.create_quad_intensity(method='mlx')
        computed_phases = [fft_shift_sim.I0, fft_shift_sim.I90, fft_shift_sim.I180, fft_shift_sim.I270]

        # compute emulated image intensity
        fft_shift_sim.create_composite_intensity(phase_variation_axis=phase_variation_axis, use_Gaussian=gauss_dim, sigma=sigma, method='mlx') # data slices
        composite_intensity = fft_shift_sim.composite_image

        # calculate the fft shift
        fft_shift_sim.calculate_fft_shifted(phase_variation_axis=phase_variation_axis)

        # reconstruct and scale intensity, phase, depth from fft shift
        recon_intensity = 2 * fft_shift_sim.reconstruct_intensity()
        recon_phase = fft_shift_sim.reconstruct_phase() - (pi/2.0)
        recon_depth = ((c * (recon_phase + pi)) / (4 * pi * mod_freq * 1e6))

        recon_intensities.append(recon_intensity)
        recon_phases.append(recon_phase)

        diff_intensities.append(intensity_rw_crop - recon_intensity)
        diff_phases.append(phase_rw_crop - recon_phase)

        snrs_intensity[index, i] = calculate_snr(intensity_rw_crop, recon_intensity, n_bucket=(0, 4))
        snrs_phase[index, i] = calculate_snr(phase_rw_crop, recon_phase, n_bucket=(0, 4))

        # Compare with n bucket reconstruction
        recon_intensity_bucket, recon_phase_bucket = fft_shift_sim.reconstruct_n_bucket(phase_variation_axis=phase_variation_axis)
        n_bucket_images.append(fft_shift_sim.n_bucket_image)
        recon_intensities_bucket.append(recon_intensity_bucket)
        recon_phases_bucket.append(recon_phase_bucket)

        diff_intensities_bucket.append(intensity_rw_crop - recon_intensity_bucket)
        diff_phases_bucket.append(phase_rw_crop - recon_phase_bucket)

        snrs_intensity_bucket[index, i] = calculate_snr(intensity_rw_crop, recon_intensity_bucket, n_bucket=(0, 0))
        snrs_phase_bucket[index, i] = calculate_snr(phase_rw_crop, recon_phase_bucket, n_bucket=(0, 0))
    
    d_index = index
    max_sigma_b_indx = np.argmax(snrs_phase_bucket[index]) 
    max_sigma_f_indx = np.argmax(snrs_phase[index]) 

    img1 = axs[d_index,0].imshow(diff_phases_bucket[max_sigma_b_indx], cmap="hsv", vmin = -pi/10, vmax=pi/10)
    axs[d_index,0].axis('off')
    
    # img2 = axs[d_index,1].imshow(phase_rw, cmap="hsv")
    img2 = axs[d_index,1].imshow(diff_phases[max_sigma_f_indx], cmap="hsv", vmin = -pi/10, vmax=pi/10)
    axs[d_index,1].axis('off')

    if d_index == 0:
        axs[d_index,0].set_title("N-bucket")
        axs[d_index,1].set_title("Fourier")
    if d_index == len(degrees)-1:
        cbar_bucket = fig_recon.colorbar(img1, ax=axs[d_index,0], shrink = 0.8, location='bottom', orientation='horizontal')
        cbar_fourier = fig_recon.colorbar(img2, ax=axs[d_index,1], shrink=0.8,location='bottom', orientation='horizontal')
        cbar_bucket.ax.tick_params(labelsize=10)
        cbar_fourier.ax.tick_params(labelsize=10)

if save_figs:
    fig_recon.savefig("figures/CVPRfigures/snapshot_n_bucket_vs_fourier-multiple_2.pdf", bbox_inches = 'tight')
```

```python
fig_snrs, ax = plt.subplots(1,1,figsize=(8.5,4), layout='constrained')
colors_bucket = plt.cm.GnBu_r(np.linspace(0.1,0.8,snrs_phase_bucket.shape[0]))
colors_fourier = plt.cm.YlOrRd_r(np.linspace(0.05,0.8,snrs_phase_bucket.shape[0]))

for i in range(snrs_phase_bucket.shape[0]):
    max_sigma_b_indx = np.argmax(snrs_phase_bucket[i]) 
    max_sigma_f_indx = np.argmax(snrs_phase[i]) 
    ax.plot(using_sigmas, snrs_phase_bucket[i], "o-", markersize=2.5, label="N bucket, k=$2\pi$/{}".format(360/degrees[i]), color=colors_bucket[i])
    ax.scatter(using_sigmas[max_sigma_b_indx], snrs_phase_bucket[i,max_sigma_b_indx],color=colors_bucket[i])
    ax.plot(using_sigmas, snrs_phase[i], "o-", markersize=2.5, label="Fourier, k=$2\pi$/{}".format(360/degrees[i]), color=colors_fourier[i])
    ax.scatter(using_sigmas[max_sigma_f_indx], snrs_phase[i,max_sigma_f_indx], color=colors_fourier[i])

ax.set_xlabel("$\sigma$")
ax.set_ylabel("SNR[dB]")
ax.set_ylim([28,40])
extent = ax.get_window_extent().transformed(fig_recon.dpi_scale_trans.inverted())
fig_snrs.legend(loc='outside right', fontsize=14)

if save_figs:
    fig_snrs.savefig("figures/CVPRfigures/snapshot_n_bucket_vs_fourier_snrs_condensed.pdf", bbox_inches = 'tight')
    
print("N-bucket SNR, R=3.5", snrs_phase_bucket[1,max_sigma_b_indx])
```

### Run simulations at k=90 and k-120 and compare results for N bucket and snapshot reconstruction

```python
k = np.deg2rad(90)
cropY = 0

intensity_rw = intensity_rw[cropX:, cropY:]
depth_rw = depth_rw[cropX:, cropY:]
phase_rw = phase_rw[cropX:, cropY:]

snrs_intensity_bucket = []
snrs_phase_bucket = []
snrs_intensity = []
snrs_phase = []

recon_intensities = []
recon_phases = []
recon_depths = []

diff_intensities = []
diff_phases = []
diff_depths = []

recon_intensities_bucket = []
recon_phases_bucket = []
recon_depths_bucket = []

diff_intensities_bucket = []
diff_phases_bucket = []
diff_depths_bucket = []

for i in range(len(using_sigmas)):
    sigma = using_sigmas[i]
       
    # create a simulation
    fft_shift_sim = fft_shift.FFT_Shift(k, 
                                        intensity_rw, depth_rw,
                                        omega= 2 * pi * (mod_freq * 1e6))

    # computationally create quad intensity measurements
    fft_shift_sim.create_quad_intensity()
    computed_phases = [fft_shift_sim.I0, fft_shift_sim.I90, fft_shift_sim.I180, fft_shift_sim.I270]

    # compute emulated image intensity
    fft_shift_sim.create_composite_intensity(phase_variation_axis=phase_variation_axis, use_Gaussian=gauss_dim, sigma=sigma) # data slices
    composite_intensity = fft_shift_sim.composite_image

    # calculate the fft shift
    fft_shift_sim.calculate_fft_shifted(phase_variation_axis=phase_variation_axis)

    # reconstruct and scale intensity, phase, depth from fft shift
    recon_intensity = 2 * fft_shift_sim.reconstruct_intensity()
    recon_phase = -fft_shift_sim.reconstruct_phase() - pi
    recon_depth = ((c * (recon_phase + pi)) / (4 * pi * mod_freq * 1e6))
    
    recon_intensities.append(recon_intensity)
    recon_phases.append(recon_phase)
    
    diff_intensities.append(intensity_rw - recon_intensity)
    diff_phases.append(phase_rw - recon_phase)
    
    snrs_intensity.append(calculate_snr(intensity_rw, recon_intensity, n_bucket=(0, 4)))
    snrs_phase.append(calculate_snr(phase_rw, recon_phase, n_bucket=(0, 4)))

    # Compare with n bucket reconstruction
    recon_intensity_bucket, recon_phase_bucket = fft_shift_sim.reconstruct_n_bucket(phase_variation_axis=phase_variation_axis)
    recon_phase_bucket = (-3*pi/2.0) - recon_phase_bucket

    recon_intensities_bucket.append(recon_intensity_bucket)
    recon_phases_bucket.append(recon_phase_bucket)
    
    diff_intensities_bucket.append(intensity_rw - recon_intensity_bucket)
    diff_phases_bucket.append(phase_rw - recon_phase_bucket)
    
    snrs_intensity_bucket.append(calculate_snr(intensity_rw, recon_intensity_bucket, n_bucket=(0, 4)))
    snrs_phase_bucket.append(calculate_snr(phase_rw, recon_phase_bucket, n_bucket=(0, 4)))
```
