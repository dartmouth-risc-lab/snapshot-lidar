```python
import numpy as np
import cv2
from scipy.constants import c, pi
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm, SymLogNorm
from scipy.ndimage import gaussian_filter1d, gaussian_filter

# uncomment for inline matlab style interactive figures
# %matplotlib widget

%cd ".."
import simulation as sim
import fft_shift as fft_shift
from utils import calculate_snr
from fig_setup import fig_setup

fig_setup() # commment this out if using matplotlib widget because they conflict
```

```python
# data paths and variables
save_figs = False

mod_freq = 60 # MHz
wavelength = 940 # nm
k_deg = 90
k = np.deg2rad(k_deg)
calib_offset = 0 # mm
cropX = 0
cropY = 0
gauss_dim='1d'
sigma=1
phase_variation_axis="x"

# from mlx datasheet
unambig_dist = (c*0.5)/(mod_freq *1e6)
coeff_to_rad = unambig_dist / (2 * pi)
```

```python
# Load average gt

with np.load('data/comparison/dist_amp_avg.npz') as gt_data:
    depth_gt = ((gt_data['dist'] - calib_offset) * 1e-3)[cropY:, cropX:] # divide by 1e3 to get meters
    intensity_gt = gt_data['amp'][cropY:, cropX:]
    
with np.load('data/comparison/phases_avg.npz') as phase_data:
    I0_gt = phase_data['phase0'][cropY:, cropX:]
    I90_gt = phase_data['phase90'][cropY:, cropX:]
    I180_gt = phase_data['phase180'][cropY:, cropX:]
    I270_gt = phase_data['phase270'][cropY:, cropX:]
    
phase_gt = np.arctan2(I0_gt - I180_gt, I270_gt - I90_gt) - (pi/2.0) # calculate the phase from the depth

with np.load('data/comparison/phases_single_frame.npz') as phase_frame_data:
    I0_frame = phase_frame_data['phase0'][cropY:, cropX:]
    I90_frame = phase_frame_data['phase90'][cropY:, cropX:]
    I180_frame = phase_frame_data['phase180'][cropY:, cropX:]
    I270_frame = phase_frame_data['phase270'][cropY:, cropX:]
```

### Simulate the same exposure duration for conventional and our methods by adding Poisson noise

```python
using_sigmas = np.linspace(0.1, 7.0, num=70)

snrs_intensity = []
snrs_phase = []
snrs_depth = []
snrs_intensity_conv = []
snrs_phase_conv = []
snrs_depth_conv = []

recon_intensities = []
recon_phases = []
recon_depths = []

diff_intensities = []
diff_phases = []
diff_depths = []

recon_intensities_conv = []
recon_phases_conv = []
recon_depths_conv = []

diff_intensities_conv = []
diff_phases_conv = []
diff_depths_conv = []

# Apply Poisson(t) to frame
t = 1000/500
I0_frame_noise = np.zeros(I0_frame.shape)
I0_frame_noise[np.where(I0_frame < 0)] = (-np.random.poisson(abs(I0_frame) * t) / (t))[np.where(I0_frame < 0)]
I0_frame_noise[np.where(I0_frame >= 0)] = (np.random.poisson(abs(I0_frame) * t) / (t))[np.where(I0_frame >= 0)]

I90_frame_noise = np.zeros(I90_frame.shape)
I90_frame_noise[np.where(I90_frame < 0)] = (-np.random.poisson(abs(I90_frame) * t) / (t))[np.where(I90_frame < 0)]
I90_frame_noise[np.where(I90_frame >= 0)] = (np.random.poisson(abs(I90_frame) * t) / (t))[np.where(I90_frame >= 0)]

I180_frame_noise = np.zeros(I180_frame.shape)
I180_frame_noise[np.where(I180_frame < 0)] = (-np.random.poisson(abs(I180_frame) * t) / (t))[np.where(I180_frame < 0)]
I180_frame_noise[np.where(I180_frame >= 0)] = (np.random.poisson(abs(I180_frame) * t) / (t))[np.where(I180_frame >= 0)]

I270_frame_noise = np.zeros(I270_frame.shape)
I270_frame_noise[np.where(I270_frame < 0)] = (-np.random.poisson(abs(I270_frame) * t) / (t))[np.where(I270_frame < 0)]
I270_frame_noise[np.where(I270_frame >= 0)] = (np.random.poisson(abs(I270_frame) * t) / (t))[np.where(I270_frame >= 0)]

I0_four = I0_frame.copy()
I90_four = I90_frame.copy()
I180_four = I180_frame.copy()
I270_four = I270_frame.copy()
quad_four = [I0_four, I90_four, I180_four, I270_four]

# Apply Poisson(4t) to frame
I0_four_noise = np.zeros(I0_frame.shape)
I0_four_noise[np.where(I0_four < 0)] = (-np.random.poisson(abs(I0_four) * t*4) / (t*4))[np.where(I0_four < 0)]
I0_four_noise[np.where(I0_four >= 0)] = (np.random.poisson(abs(I0_four) * t*4) / (t*4))[np.where(I0_four >= 0)]

I90_four_noise = np.zeros(I90_frame.shape)
I90_four_noise[np.where(I90_four < 0)] = (-np.random.poisson(abs(I90_four) * t*4) / (t*4))[np.where(I90_four < 0)]
I90_four_noise[np.where(I90_four >= 0)] = (np.random.poisson(abs(I90_four) * t*4) / (t*4))[np.where(I90_four >= 0)]

I180_four_noise = np.zeros(I180_four.shape)
I180_four_noise[np.where(I180_four < 0)] = (-np.random.poisson(abs(I180_four) * t*4) / (t*4))[np.where(I180_four < 0)]
I180_four_noise[np.where(I180_four >= 0)] = (np.random.poisson(abs(I180_four) * t*4) / (t*4))[np.where(I180_four >= 0)]

I270_four_noise = np.zeros(I270_frame.shape)
I270_four_noise[np.where(I270_four < 0)] = (-np.random.poisson(abs(I270_four) * t*4) / (t*4))[np.where(I270_four < 0)]
I270_four_noise[np.where(I270_four >= 0)] = (np.random.poisson(abs(I270_four) * t*4) / (t*4))[np.where(I270_four >= 0)]

for i in range(len(using_sigmas)):
    sigma = using_sigmas[i]
      
    ######### create a snapshot simulation #########
    fft_shift_sim = fft_shift.FFT_Shift(k,
                                        intensity_gt, depth_gt,
                                        omega = 2.0 * pi * (mod_freq * 1e6))

    # computationally create quad intensity measurements
    fft_shift_sim.create_quad_intensity_from_data(quad_four)
    computed_phases = [fft_shift_sim.I0, fft_shift_sim.I90, fft_shift_sim.I180, fft_shift_sim.I270]
    
    # compute emulated image intensity
    fft_shift_sim.create_composite_intensity_from_data(images=[I0_four_noise, I90_four_noise, I180_four_noise, I270_four_noise], sigma=sigma)
    composite_intensity = fft_shift_sim.composite_image

    # calculate the fft shift
    fft_shift_sim.calculate_fft_shifted(phase_variation_axis=phase_variation_axis)

    # reconstruct and scale intensity, phase, depth from fft shift
    recon_intensity = fft_shift_sim.reconstruct_intensity()
    recon_phase = -fft_shift_sim.reconstruct_phase()
    
    recon_intensities.append(recon_intensity)
    recon_phases.append(recon_phase)
    
    diff_intensities.append(intensity_gt - recon_intensity)
    diff_phases.append(phase_gt - recon_phase)
    
    snrs_intensity.append(calculate_snr(intensity_gt, recon_intensity))
    snrs_phase.append(calculate_snr(phase_gt, recon_phase))

    ######### create a conventional, quad bin simulation #########
    I0_frame_noise_cur = gaussian_filter(I0_frame_noise, sigma, mode='constant')
    I90_frame_noise_cur = gaussian_filter(I90_frame_noise, sigma, mode='constant')
    I180_frame_noise_cur = gaussian_filter(I180_frame_noise, sigma, mode='constant')
    I270_frame_noise_cur = gaussian_filter(I270_frame_noise, sigma, mode='constant')
    intensity_conventional = 2.0 * fft_shift_sim.compute_amplitude_conventional(images=[I0_frame_noise_cur, I90_frame_noise_cur, I180_frame_noise_cur, I270_frame_noise_cur])
    phase_conventional = fft_shift_sim.compute_phase_conventional(images=[I0_frame_noise_cur, I90_frame_noise_cur, I180_frame_noise_cur, I270_frame_noise_cur])
    
    recon_intensities_conv.append(intensity_conventional)
    recon_phases_conv.append(phase_conventional)
    
    diff_intensities_conv.append(intensity_gt - intensity_conventional)
    diff_phases_conv.append(phase_gt - phase_conventional)
    
    snrs_intensity_conv.append(calculate_snr(intensity_gt, intensity_conventional))
    snrs_phase_conv.append(calculate_snr(phase_gt, phase_conventional))
```

```python
# Display snr for optimal sigma=1
print("SNR Phase Conventional: ", snrs_phase_conv[9])
print("SNR Phase Snapshot: ", snrs_phase[9])
```
